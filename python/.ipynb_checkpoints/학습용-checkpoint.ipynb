```python
cb_data = pd.read_csv('cb_data.csv')
cb_data
```

```python
tr_new = pd.read_csv('transaction_new.csv')
tr_new.head()
```

```python
tr_new.shape
```

```python
tr_new_unique = tr_new.drop_duplicates(subset='product_id')

cb_data_merge = pd.merge(cb_data, tr_new_unique[['product_id','item_price']], on='product_id', how = 'left')
cb_data_merge
```

```python
cb_data_merge=cb_data_merge[['product_id','quantity', 'masterCategory','articleType','productDisplayName','item_price']]
```

```python
cb_product = cb_data_merge.groupby(['product_id','masterCategory','articleType','productDisplayName','item_price'])['quantity'].count().reset_index()
cb_product
```

```python
cb_product.to_csv('cb_product.csv', index = False)
```
