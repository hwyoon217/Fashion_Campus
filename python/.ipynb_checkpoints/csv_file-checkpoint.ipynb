```python
expected_fields = 10  # 정상적인 컬럼 개수로 변경하세요

with open('product.csv', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f, 1):
        fields = line.strip().split(',')
        if len(fields) != expected_fields:
            print(f'Line {i} has {len(fields)} fields: {line[:200]}')  # 일부만 출력
```

```python
corrected_lines = []

expected_fields = 10
with open('product.csv', 'r', encoding='utf-8') as f:
    header = f.readline().strip()
    corrected_lines.append(header)

    for i, line in enumerate(f, 2):
        fields = line.strip().split(',')

        if len(fields) >= expected_fields:
            # 앞 9개 필드 + 나머지를 모두 합치고, 큰따옴표로 감싸기
            product_name = ','.join(fields[9:]).strip('"')
            product_name_quoted = f'"{product_name}"'
            fixed_fields = fields[:9] + [product_name_quoted]
            corrected_lines.append(','.join(fixed_fields))
        else:
            print(f"⚠️ Line {i} has too few fields: {len(fields)} fields")

# 새 파일로 저장
with open('product_fixed.csv', 'w', encoding='utf-8') as f:
    for line in corrected_lines:
        f.write(line + '\n')

print("✅ productDisplayName 필드를 큰따옴표로 감싸서 저장 완료")
```
